```python
text = "Pokémon, short for 'Pocket Monsters,' is an enduring global phenomenon that originated in Japan in 1996 with the release of the first video games by Nintendo, Game Freak, and Creatures. The franchise quickly expanded into trading card games, an animated television series, movies, toys, and a wide array of merchandise. At its core, Pokémon follows the adventures of trainers who capture, train, and battle creatures known as Pokémon, each possessing unique abilities and characteristics. The primary objective is to become a Pokémon Master by collecting and battling with various Pokémon, aiming to defeat powerful opponents and gym leaders to earn badges and prestige. Central to the franchise's appeal is its blend of strategic gameplay and an emotional connection with the creatures, exemplified by iconic characters like Pikachu, Charizard, and legendary Pokémon such as Mewtwo. The animated series, notably featuring Ash Ketchum and his steadfast companion Pikachu, has captivated audiences for decades, showcasing themes of friendship, perseverance, and teamwork. Pokémon’s cultural impact is immense, fostering community-driven events like Pokémon GO meetups and tournaments, which emphasize social interaction and exploration. With new generations of games and innovations, Pokémon continues to evolve, captivating fans of all ages and cementing its status as a cherished and nostalgic part of pop culture."
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
#mapping each token to a number
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])
tokenizer.word_index
```

```python
#printing each sentence in the text
for sentence in text.split('.'):
  print(sentence)
```

```python
#converting each sentence to a vector
for sentence in text.split('.'):
  print(tokenizer.texts_to_sequences([sentence])[0])
```

```python
#need to make a dataset out of these vectors so as to make a model to predict next word
input_sequences = []
for sentence in text.split('.'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]
  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
input_sequences
```

```python
#need to add padding to make the length equal of each sequence
max_len = max(len(x) for x in input_sequences)
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences,maxlen=max_len,padding='pre')
padded_input_sequences
```

```python
#now from the input_sequences we can make the last column as output and remaining words as input for each row
X = padded_input_sequences[:,:-1]
y = padded_input_sequences[:,-1]
```

```python
X.shape , y.shape
```

```python
#need to OHE the output column
from tensorflow.keras.utils import to_categorical
encoded_y = to_categorical(y,num_classes=len(tokenizer.word_index)+1)
```

```python
encoded_y.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,LSTM,Dense
```

```python
from ast import mod
#creating the model
model = Sequential()
model.add(Embedding(input_dim=len(tokenizer.word_index)+1,output_dim=100))
model.add(LSTM(128))
model.add(Dense(len(tokenizer.word_index)+1,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(x=X , y=encoded_y , epochs=100)
```

```python
#need to find the output in the form of vector
input_text = "ash"
token_input = tokenizer.texts_to_sequences([input_text])[0]
padded_token_input = pad_sequences([token_input],maxlen=max_len-1,padding='pre')
predicted_output = model.predict(padded_token_input)
predicted_output
```

```python
#above was a vector with each component being the probabilty against that word
#so max probability word is the output we want so need to find the index
import numpy as np
max_index = np.argmax(predicted_output)
max_index
```

```python
for (word , index) in tokenizer.word_index.items():
  if index == max_index:
    print(word)
```

```python
#lets check for more than few words
input_text = "ash"
for i in range(max_index):
  token_input = tokenizer.texts_to_sequences([input_text])[0]
  padded_token_input = pad_sequences([token_input],maxlen=max_len-1,padding='pre')
  max_index = np.argmax(model.predict(padded_token_input))
  for (word , index) in tokenizer.word_index.items():
    if index == max_index:
      input_text += " " + word
      print(input_text)
```

```python
input_text = "mewtwo"
for i in range(max_index):
  token_input = tokenizer.texts_to_sequences([input_text])[0]
  padded_token_input = pad_sequences([token_input],maxlen=max_len-1,padding='pre')
  max_index = np.argmax(model.predict(padded_token_input))
  for (word , index) in tokenizer.word_index.items():
    if index == max_index:
      input_text += " " + word
      print(input_text)
```
